```python
# removing irrelevant columns
columns_to_drop = [
    'short_name', 'long_name','player_id', 'player_url', 'player_id', 'player_face_url','fifa_version', 'fifa_update', 'fifa_update_date', 'age', 'dob', 'height_cm', 'weight_kg',
    'league_id', 'league_name', 'league_level', 'club_team_id', 'club_name', 'club_position',
    'club_jersey_number', 'club_joined_date', 'club_contract_valid_until_year', 'nationality_id',
    'nationality_name', 'preferred_foot', 'weak_foot', 'skill_moves', 'international_reputation',
    'work_rate', 'body_type', 'real_face'
]

# Drop the identified columns from your dataframe
df.drop(columns=columns_to_drop,inplace=True)
```

```python
# Removing the columns with more than 30% na values
L = []
L_less = []
for i in df.columns:
    if ((df[i].isnull().sum() < (0.3*(df.shape[0])))):
        L.append(i)
    else:
        L_less.append(i)
        
print(L)
print(L_less)

df.drop(L_less, axis=1, inplace=True)
```

```python
df.shape
```

```python
#splitting the data between numeric and categorial data 
numeric_data = df.select_dtypes(include=np.number)
non_numeric= df.select_dtypes(include=['object'])
```

Working with numeric data

```python
numeric_data.info()
```

```python
# imputing the numeric_data for some of the missing values
from sklearn.impute import SimpleImputer
sc = SimpleImputer(strategy='median')
scaled = sc.fit_transform(numeric_data)
numeric_data = pd.DataFrame(scaled, columns=numeric_data.columns)
numeric_data.info()
```

```python
numeric_data.info()
```

```python
# which features are more correlated with the overall rating
corr_matrix = numeric_data.corr()
corr_matrix["overall"].sort_values(ascending=False)
```

```python
numeric_data = numeric_data[['overall', 'movement_reactions', 'mentality_composure','potential',
                            'wage_eur','value_eur','passing','attacking_short_passing']]
```

Working with non numeric data

```python
non_numeric.info()
```

```python
# Example assuming `non_numeric` contains your categorical variables
encoded_non_numeric = non_numeric.copy()

# Calculate mean target value for each category
for col in encoded_non_numeric.columns:
    target_means = df.groupby(col)['overall'].mean()
    encoded_non_numeric[col] = encoded_non_numeric[col].map(target_means)

encoded_non_numeric
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestRegressor
# Initialize RandomForestRegressor
rf = RandomForestRegressor(random_state=42)
rf.fit(encoded_non_numeric, df['overall'])

# Feature selection based on importance
selector = SelectFromModel(rf, prefit=True)
selected_features = selector.get_support(indices=True)
selected_feature_names = encoded_non_numeric.columns[selected_features]

# Subset DataFrame with selected features
encoded_non_numeric_selected = encoded_non_numeric[selected_feature_names]
```

```python
encoded_non_numeric_selected
```

```python
# Concatenate numeric_data and non_numeric
combined_df = pd.concat([numeric_data, encoded_non_numeric_selected], axis=1)
```

```python
# splitting the data into dependent and independent variables 
X = combined_df.drop('overall', axis=1)
y = combined_df['overall']
```

```python
X.shape
```

```python
y.shape
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

scaled=sc.fit_transform(X)

scaled
```

# Training the model

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,y,test_size=0.2,random_state=50)

print(Xtrain.shape)

Ytest.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import ElasticNet, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
```

```python
dtree = DecisionTreeRegressor(max_depth=12)
ridge = Ridge()
lasso = Lasso()
ENet = ElasticNet()
gradientDescent = SGDRegressor()
l = LinearRegression()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_squared_log_error, r2_score
```

```python
import pickle as pkl
```

```python
for model in (dtree, ridge, lasso, ENet, gradientDescent, l):
    model.fit(Xtrain, Ytrain)
    pkl.dump(model, open('/Users/stanleywebs911/Documents/Intro-AI/StanleyTulaniNdlovu_SportsPrediction/' + model.__class__.__name__+ '.pkl', 'wb'))
    y_pred = model.predict(Xtest)
    print(f"""{model.__class__.__name__}, Mean Absolute Error={mean_absolute_error(y_pred,Ytest)},
            Mean Square Error={mean_squared_error(y_pred,Ytest)},
            Root Mean Squared Error={np.sqrt(mean_squared_error(y_pred,Ytest))},
            R2_Score= {r2_score(y_pred,Ytest)}
 
 """)
```
